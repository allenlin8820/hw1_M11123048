```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('C:/Users/liping/OneDrive/文件/學校資料/碩一上/資料探勘/adult.data.csv', na_values = '?', names = ['age', 'workclass', 'fnlwgt', 'education', 
                                                                                'education-num','marital-status', 'occupation', 
                                                                                'relationship', 'race', 'sex','capital-gain', 
                                                                                'capital-loss', 'hours-per-week', 'native-country',
                                                                                'income'] )
dataset2 = pd.read_csv('C:/Users/liping/OneDrive/文件/學校資料/碩一上/資料探勘/adult.test.csv', na_values = '?', names = ['age', 'workclass', 'fnlwgt', 'education', 
                                                                                'education-num','marital-status', 'occupation', 
                                                                                'relationship', 'race', 'sex','capital-gain', 
                                                                                'capital-loss', 'hours-per-week', 'native-country',
                                                                                'income'] )
```

```python
df = dataset
df = df.replace('?', np.NaN)
df = df.dropna()
df
```

```python
df2 = dataset2
df2 = df2.replace('?', np.NaN)
df2 = df2.dropna()
df2
```

```python
#資料轉換(one-hot)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df[['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', 'income']]: 
    df[col] = le.fit_transform(df[col])

df
```

```python
le2 = LabelEncoder()

for col in df2[['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', 'income']]: 
    df2[col] = le2.fit_transform(df2[col])

df2
```

```python
#分割資料集

X = df.drop(['income'], axis=1) #刪除欄位
Y = df['income'] 
x = df2.drop(['income'], axis=1)
y = df2['income']
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier(criterion='gini',max_depth = 7)
clf = clf.fit(X, Y)
```

```python
#預測
test_predict = clf.predict(x)
print(test_predict)
print(Y)
```

```python
clf.score(X, Y)
```

```python
clf.score(x, y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(100, 50))
fn = ['age', 'workclass', 'fnlwgt', 'education', 'education-num','marital-status', 'occupation', 'relationship', 
      'race', 'sex','capital-gain', 'capital-loss', 'hours-per-week', 'native-country']
tree.plot_tree(clf, feature_names = fn)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt  

score_array=[]
score_array_test=[]
for i in range(1,15):
    clf = tree.DecisionTreeClassifier(criterion='gini', max_depth = i)
    clf.fit(X,Y)
    score_val = clf.score(X,Y)
    score_test = clf.score(x,y)
    score_array.append(score_val)
    score_array_test.append(score_test)

print("score_array",score_array)
print("score_array_MAX:",np.max(score_array))
print("-----------------------------")
print("score_array_test",score_array_test)
print("score_array_test_MAX:",np.max(score_array_test))

plt.figure(figsize=(10,4))
plt.xticks(range(15))  # 設定x刻度
plt.title('Gini : Max depth & Acc')
plt.grid(True,linestyle = "--",color = 'gray' ,linewidth = '0.5',axis='both')
plt.plot(score_array, label='Training Accuracy') # Plot training error over domain
plt.plot(score_array_test, label='Testing Accuracy') # Plot testing error over domain
plt.xlabel('Max depth') # Label x-axis
plt.ylabel('Accuracy') # Label y-axis
plt.legend(loc = 'upper left')
plt.show()
print(np.shape(X))
score_array=[]
score_array_test=[]
```
