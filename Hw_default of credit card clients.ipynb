```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('D:/Datamining/default of credit card clients.csv', na_values = '?' )
```

```python
df = dataset
df = df.replace('?', np.NaN)
df = df.dropna()
df
```

```python
#分割資料集
#x_train = df.sample(frac=0.7)
#x_test= df[~df.index.isin(x_train.index)]
X = df.drop(['default payment next month'], axis=1) #刪除欄位
Y = df['default payment next month'] 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt  

score_array=[]
score_array_test=[]
for i in range(1,15):
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth = i)
    clf.fit(x_train,y_train)
    score_val = clf.score(x_train,y_train)
    score_test = clf.score(x_test,y_test)
    score_array.append(score_val)
    score_array_test.append(score_test)

print("score_array",score_array)
print("score_array_MAX:",np.max(score_array))
print("-----------------------------")
print("score_array_test",score_array_test)
print("score_array_test_MAX:",np.max(score_array_test))

plt.figure(figsize=(10,4))
plt.xticks(range(15))  # 設定x刻度
plt.title('entropy : Max depth & Acc')
plt.grid(True,linestyle = "--",color = 'gray' ,linewidth = '0.5',axis='both')
plt.plot(score_array, label='Training Accuracy') # Plot training error over domain
plt.plot(score_array_test, label='Testing Accuracy') # Plot testing error over domain
plt.xlabel('Max depth') # Label x-axis
plt.ylabel('Accuracy') # Label y-axis
plt.legend(loc = 'upper left')
plt.show()
print(np.shape(x_train))
score_array=[]
score_array_test=[]
```
